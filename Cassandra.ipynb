```python
from cassandra.cluster import Cluster
import cassandra
import numpy as np
```

```python
cluster = Cluster(['172.17.0.2'])
```

```python
session = cluster.connect('twitter')
```

```python
session.set_keyspace('twitter')
```

```python
#rows = session.execute("SELECT * FROM tweet WHERE user='hablemosdelarse'")
#rows = session.execute('SELECT * FROM tweet')
rows = session.execute("SELECT * FROM tweet;")
for user_row in rows:
    print(user_row.text);
    print("---")
```

```python
def to_integer(dt_time):
    return str(10000*dt_time.year)[0:4]+ "-" + str(100*dt_time.month)[0:1] + "-" + str(dt_time.day)
```

```python
import tweepy
consumer_key=''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

public_tweets = api.home_timeline(count=100)
for tweet in public_tweets:
    session.execute("INSERT INTO tweet (user, text, date) VALUES ('{}', '{}', '{}');".format(tweet.user.screen_name,tweet.text.replace("'",""),to_integer(tweet.created_at)))
```
